```python
# your code goes here
import pandas as pd
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv"
df = pd.read_csv(url)
df.columns = list(map(lambda x : x.lower().replace(" ","_"), df.columns))
df.columns
```

```python
df
```

```python
# 1. Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" 
# to the last marketing campaign.

df["claim2"] = df["total_claim_amount"].apply(lambda x: "high claim" if x > 1000 else "low claim")
new_df = df.groupby("claim2").get_group('high claim')
new_df = new_df.groupby('response').get_group('Yes')
new_df
```

```python
#2. Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded 
#"Yes" to the last marketing campaign. Write your conclusions.

df_yes = df.groupby('response').get_group('Yes') ##creates a df that only considers customers that responded Yes
df_yes.groupby(['policy_type','gender'])["total_claim_amount"].mean().round(2)

#averages are relatively similar between the groups
```

```python
#3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where 
#there are more than 500 customers.

df['customer_count'] = df.groupby(['state','policy'])["policy"].transform('count')
filtered_states = df[df['customer_count']>500]
filtered_states.groupby(['state','policy'])["policy"].count()
```

```python
#4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.
print("")
print("MAX")
print(df.groupby(['gender','education'])["customer_lifetime_value"].max().round(2))
print('')
print("MIN")
print(df.groupby(['gender','education'])["customer_lifetime_value"].min().round(2))
print('')
print("MEDIAN")
print("")
print(df.groupby(['gender','education'])["customer_lifetime_value"].median().round(2))

print(df.groupby(['gender','education'])["customer_lifetime_value"].agg(['max','min','median']))
```

```python
df.groupby(['education', 'gender'])['customer_lifetime_value'].agg(['max', 'min', 'median'])
```

```python
# 5. The marketing team wants to analyze the number of policies sold by state and month. Present the data in a table where the months are 
# arranged as columns and the states are arranged as rows.
#not sure what to use as month here

policy_count_by_state_month = df.groupby(['state', 'month']).size().reset_index(name='policy_count')
policy_count_pivot = policy_count_by_state_month.pivot(index='state', columns='month', values='policy_count')
print(policy_count_pivot)
```

```python
# 6. Display a new DataFrame that contains the number of policies sold by month, by state, for the top 3 states with the highest number of policies sold.

# *Hint:*
# - *To accomplish this, you will first need to group the data by state and month, then count the number of policies sold for each group. Afterwards, you will need to sort the data by the count of policies sold in descending order.*
# - *Next, you will select the top 3 states with the highest number of policies sold.*
# - *Finally, you will create a new DataFrame that contains the number of policies sold by month for each of the top 3 states.*
```

```python
policy_count_by_state_month

##sorting the data

policy_count_by_state_month_sorted = policy_count_by_state_month.sort_values(by='policy_count', ascending=False)

# Select the top 3 states with the highest number of policies sold
top_states = policy_count_by_state_month_sorted['state'].unique()[:3]

# Create a new DataFrame that contains the number of policies sold by month for each of the top 5 states
top_states_policies = policy_count_by_state_month_sorted[policy_count_by_state_month_sorted['state'].isin(top_states)]
top_states_policies
```

```python
# 7. The marketing team wants to analyze the effect of different marketing channels on the customer response rate.

# Hint: You can use melt to unpivot the data and create a table that shows the customer response rate (those who responded "Yes") 
# by marketing channel.

df['response'] = df['response'].apply(lambda x: 1 if x=='Yes' else 0) #Transform the Yes and No into integers
melted_table = pd.melt(df, id_vars=['sales_channel'], value_vars=['response'])
melted_table
```

```python
grouped_table = melted_table.groupby('sales_channel').value.value_counts().unstack().fillna(0)
grouped_table
```
